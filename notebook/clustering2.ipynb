# 데이터 요인 선택

```r
# 최적화된 군집화 찾기 
df_1<-select(df2,'Q5_1','CQ14_1T','D_COU','D_GEN')
df_2<-select(df2,'Q5_1','CQ14_1T','D_COU','D_AGE','D_GEN','Q16_a')
df_3<-select(df2,'Q2A1','Q5_1','CQ14_1T','D_COU','D_GEN','Q16_a')
df_4<-select(df2,'Q2A1','Q5_1','CQ14_1T','D_COU','D_GEN')
df_5<-select(df2,'Q2A1','Q5_1','CQ14_1T','D_COU','D_GEN','D_AGE','Q16_a')
```

```r
# 정규화
df_s1 <- scale(df_1)
df_s2 <- scale(df_2)
df_s3 <- scale(df_3)
df_s4 <- scale(df_4)
df_s5 <- scale(df_5)
```

```r
# 군집 갯수(k)추천
fviz_nbclust(df_s1,kmeans,method='gap_stat')
fviz_nbclust(df_s2,kmeans,method='gap_stat')
fviz_nbclust(df_s3,kmeans,method='gap_stat')
fviz_nbclust(df_s4,kmeans,method='gap_stat')
fviz_nbclust(df_s5,kmeans,method='gap_stat')
```

```r
# df_s1 : k = 5 / 68.2 %
# df_s2 : k = 6 / 56.9 %
# df_s3 : k = 5 / 53.1 %
# df_s4 : k = 2 / 24.4 %
# df_s5 : k = 2 / 17.5 %
```

```r
# df_s1 : k = 5 / 68.2 %
residual1 <-kmeans(df_s1,5,nstart = 25)
residual1
fviz_cluster(residual1,data = df_s1)
```

```r
# df_s2 : k = 6 / 56.9 %
residual2 <-kmeans(df_s2,6,nstart = 25)
residual2
fviz_cluster(residual2,data = df_s2)
```

```r
# df_s3 : k = 5 / 53.1 %

residual3 <-kmeans(df_s3,5,nstart = 25)
residual3
fviz_cluster(residual3,data = df_s3)
```

```r
# df_s4 : k = 2 / 24.4 %

residual4 <-kmeans(df_s4,2,nstart = 25)
residual4
fviz_cluster(residual4,data = df_s4)
```

```r
# df_s5 : k = 2 / 17.5 %

residual5 <-kmeans(df_s5,2,nstart = 25)
residual5
fviz_cluster(residual5,data = df_s5)
```
